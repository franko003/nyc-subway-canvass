```python
# https://data.cityofnewyork.us/Education/DOE-High-School-Directory-2017/s3k6-pzi2

import pandas as pd

all_high_schools = pd.read_csv('data/DOE_High_School_Directory_2017.csv')
all_high_schools.shape
```

```python
pd.set_option('display.max_columns', 453)
all_high_schools.head(3)
```

```python
boys_only = all_high_schools['boys'] == 1
columns_of_interest = ['dbn', 'school_name', 'boro', 'academicopportunities1', 'academicopportunities2', 
                       'academicopportunities3', 'academicopportunities4', 'academicopportunities5',
                       'neighborhood', 'location', 'subway', 'bus', 'total_students', 'start_time', 'end_time',
                       'graduation_rate', 'attendance_rate', 'pct_stu_enough_variety', 'college_career_rate', 
                       'girls', 'specialized', 'earlycollege', 'program1', 'program2', 'program3', 'program4', 
                       'program5', 'program6', 'program7', 'program8', 'program9', 'program10', 'interest1',
                       'interest2', 'interest3', 'interest4', 'interest5', 'interest6', 'interest7', 'interest8', 
                       'interest9', 'interest10', 'city', 'zip']
df = all_high_schools[~boys_only][columns_of_interest]
df.set_index('dbn', inplace=True)
df.shape
```

```python
df.head(3)
```

```python
import numpy as np

def contains_terms(column_name, terms=["tech"]):
    contains = []
    for i, term in enumerate(terms):
        contains.append(df[column_name].str.contains(terms[i], case=False))

    not_null = df[column_name].notnull()
    return (not_null) & (np.any(contains, axis=0))
    

def contains_terms_columns(column_root, n_columns, terms=["tech"]):
    if n_columns == 1:
        return contains_terms(column_root, terms)
    
    tech = []
    for i in range(n_columns):
        column_name = column_root + str(i + 1)
        tech.append(contains_terms(column_name, terms))
    
    return np.any(tech, axis=0)
```

```python
tech_academicopportunities = contains_terms_columns('academicopportunities', 5, 
                                                    terms=['technology', 'computer', 'web', 'programming', 'coding'])
len(df[tech_academicopportunities])
```

```python
# searching for 'tech' might match the word 'technical' 
all_tech_program = contains_terms_columns('program', 10, terms=['programming', 'computer', 'tech'])
technical_program = contains_terms_columns('program', 10, terms=['technical'])
tech_program = (all_tech_program) & ~(technical_program)
len(df[tech_program])
```

```python
tech_interest = contains_terms_columns('interest', 10, terms=['computer', 'technology'])
len(df[tech_interest])
```

```python
tech_inclined = (tech_academicopportunities) | (tech_program) | (tech_interest)
print(len(df[tech_inclined]))
print("{:.1f}%".format(100 * len(df[tech_inclined]) / len(df)))
```

```python
df['tech_academicopportunities'] = tech_academicopportunities.astype(int)
df['tech_program'] = tech_program.astype(int)
df['tech_interest'] = tech_interest.astype(int)
df.head(3)
```

```python
new_columns = ['school_name', 'boro', 'tech_academicopportunities', 'neighborhood', 'location', 'subway', 
               'bus', 'total_students', 'start_time', 'end_time', 'graduation_rate', 'attendance_rate', 
               'pct_stu_enough_variety', 'college_career_rate', 'girls', 'specialized', 'earlycollege',
               'tech_program', 'tech_interest', 'city', 'zip']
tech_schools = df[tech_inclined][new_columns]
tech_schools.head(3)
```

```python
potential = tech_schools['college_career_rate'] * tech_schools['college_career_rate']
potential.sort_values(inplace=True, ascending=False)
potential
```

```python
null_college_career_rate = tech_schools.college_career_rate.isnull()
print("{:.1f}%".format(100 * len(tech_schools[null_college_career_rate]) / len(tech_schools)))
```

```python
null_graduation_rate = tech_schools.graduation_rate.isnull()
print("{:.1f}%".format(100 * len(tech_schools[null_graduation_rate]) / len(tech_schools)))
```

```python
print("{:.1f}%".format(100 * \
                       len(tech_schools[(null_college_career_rate) & (null_graduation_rate)]) / len(tech_schools)))
```

```python
pd.options.mode.chained_assignment = None

tech_college_potential = tech_schools.dropna(subset=['college_career_rate'])
tech_college_potential['potential_college_students'] = (tech_college_potential['total_students'] * \
                                                        tech_college_potential['college_career_rate']).astype(int)

tech_college_potential.sort_values('potential_college_students', inplace=True, ascending=False)
tech_college_potential.head(3)
```

```python
tech_college_potential['potential_college_students'].hist(bins=range(0, 6000, 250), 
                                                          figsize=(12, 5))
```

```python
high_potential = tech_college_potential['potential_college_students'] > 1000
high_potential_schools = tech_college_potential[high_potential]
len(high_potential_schools)
```

```python
high_potential_schools.loc[:, ('subway', 'bus')]
```

```python
close_to_subway = high_potential_schools.dropna(subset=['subway'])
close_to_subway
```

```python
def add_coord_columns(dataframe, condensed_column, sep=', '):
    coord_str = dataframe.apply(lambda x: x[condensed_column][x[condensed_column].find("(") + 1 : \
                                                              x[condensed_column].find(")")], \
                                axis='columns')

    dataframe['latitude'] = coord_str.apply(lambda x: x[ : x.find(sep)]).astype(float)
    dataframe['longitude'] = coord_str.apply(lambda x: x[x.find(sep) + 1 : ]).astype(float)
```

```python
add_coord_columns(close_to_subway, 'location')
close_to_subway.loc[:, ('latitude', 'longitude')]
```

```python
close_to_subway.rename(columns={'subway': 'subway_lines'}, inplace=True)

df_to_pickle = close_to_subway.loc[:, ('school_name', 'potential_college_students', 'latitude', 'longitude', 
                                    'start_time', 'end_time', 'subway_lines')]
df_to_pickle
```
